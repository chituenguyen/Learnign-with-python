### pyplot api vs oo api
* pyplot -> quickly
* oo api -> advanced 

```python
# pyplot api
x=np.linspace(0,10,1000) # 0 den 10 1000 cai 
x
```

```python
plt.title("Sinx and Cosx")
plt.plot(x,np.sin(x),color='yellow',linestyle="dashed",label="sinx")
plt.plot(x,np.cos(x),color="red",linestyle="dashed",label="cosx")
plt.xlabel('x')
plt.ylabel('fx')
# plt.xlim([0,4])
# plt.axis([xmin,xmax,ymin,ymax])
plt.axis("equal")
plt.legend()
plt.show()
```

### oo api


```python
#1 Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2 Set up plot
fig,ax=plt.subplots(figsize=(5,5))
# PLot data
ax.plot(x,y)
ax.set(title="Simple plot",
        xlabel="x-axis",
        ylabel="y-axis")
plt.show()
```

### Most common types of Matplotlib
* Line
* Scatter
* Bar
* Hist
* Subplots

```python
# Line 
x=np.linspace(0,10,100)
x[:5]
```

```python
# oo api
fig,ax = plt.subplots()
ax.plot(x,x**3)
plt.show()
```

### Scatter


```python
# pyplot api
plt.scatter(x,np.exp(x))
plt.show()
```

```python
# oo api
rng=np.random.RandomState(0)
x=rng.randn(100)
x
```

```python
y=rng.randn(100)
```

```python
sizes=100 * rng.rand(100)
color=rng.rand(100)
fig,ax=plt.subplots()
img1 = ax.scatter(x,y,s=sizes,c=color,cmap="viridis",alpha=0.5)
fig.colorbar(img1)
plt.show()
```

### Bar
    * Vertical
    * Horizontal

```python
# Prepare data
soft_drink_prices = {"Coca":10,
                     "Pepsi":12,
                     "Number":15}
fig,ax = plt.subplots()
ax.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax.set(title="Bach hoa xanh",ylabel="Price ($)")
plt.show()
```

```python
fig,ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()),list(soft_drink_prices.values()))
plt.show()
```

### histogram


```python
np.random.seed(42)
student_height = np.random.normal(170,10,250) #Lo (mean) do lech chuan 10 250 hoc sinh
student_height
```

```python
plt.hist(student_height,bins=30)
plt.show()
```

### Subplots

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax1.plot(x,x/2)
ax1.plot(x,x+2)

ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax4.hist(student_height,bins=30)
plt.show()
```
