```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
data
```

```python
data = pd.read_csv('train.csv', index_col = "Id")
data
```

```python
data.columns
```

### Using feature engineering to choose Feature needed

```python
features = ["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd"]
```

```python
# split dataset
# x: data[feature]
# y: data[saleprice]
```

```python
X = data[features]
X
```

```python
y = data['SalePrice']
y
```

### train, valid

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_valid,y_train,y_valid = train_test_split(X,y,train_size=0.8, test_size=0.2,random_state=0)
X_train
```

### Training MC

```python
from sklearn.tree import DecisionTreeRegressor



dt_model = DecisionTreeRegressor(random_state=1)
```

```python
# Fit training data into model
dt_model.fit(X_train,y_train)
```

```python
y_predict = dt_model.predict(X_valid.head())
```

```python
y_predict
```

```python
pd.DataFrame({'y':y_valid.head(),'y_predict':y_predict})
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X_train,y_train)
rf_val_preds = rf_model.predict(X_valid)
```

```python
pd.DataFrame({'y':y_valid.head(),'y_predict':rf_val_preds[:5]})
```
