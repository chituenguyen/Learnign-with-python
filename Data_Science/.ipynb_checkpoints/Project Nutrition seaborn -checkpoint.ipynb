```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
cereal_df = pd.read_csv('nutrition.csv')
%matplotlib inline
```

```python
cereal_df #mfr nha san xuat, type co 2 loai C and H 
```

```python
# drop fields
fields=['shelf','weight','cups','rating']
cereal_df_new=cereal_df.drop(fields,axis=1)
```

```python
cereal_df_new
```

```python
# corr()
cereal_corr = cereal_df_new.corr() # get correlation data
```

```python
cereal_corr
```

```python
# ones_like build a mattrix of booleans with the same shape as our data
ones_corr = np.ones_like(cereal_corr,dtype=bool)
```

```python
ones_corr
```

```python
# np.triu return upper triangle mattrix 
mask = np.triu(ones_corr)
```

```python
mask
```

```python
sns.heatmap(data=cereal_corr,mask=mask)
```

```python
adjust_mask = mask[1:,:-1]
```

```python
adjust_cereal_corr = cereal_corr.iloc[1:,:-1]
```

```python
fig, ax = plt.subplots(figsize=(10,8))
cmap=sns.diverging_palette(0,230,90,60,as_cmap=True) # hsluv.org
sns.heatmap(data=adjust_cereal_corr,mask=adjust_mask,annot=True,fmt=".2f",cmap=cmap,vmin=-1,vmax=1,linewidths=0.5)
yticks = [i.upper()for i in adjust_cereal_corr.index]
xticks = [i.upper() for i in adjust_cereal_corr.columns]
ax.set_yticklabels(yticks,rotation=0)
ax.set_xticklabels(xticks,rotation=90,fontsize=10)
ax.set_title("Correlation matrix cereal")
```
